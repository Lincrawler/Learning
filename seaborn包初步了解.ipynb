```python
import pandas
iris=pandas.read_csv('iris.csv')
```

```python
iris.sample(10)
```

```python
import seaborn
%matplotlib inline
seaborn.countplot("species",data=iris)
```

```python
seaborn.barplot(x='species',y='petal_length',data=iris)
```

```python
seaborn.boxplot(x='species',y='petal_length',data=iris)
```

```python
seaborn.distplot(iris['petal_length'])
```

```python
iris_vir=iris[iris.species=='Iris-virginica']
iris_ver=iris[iris.species=='Iris-versicolor']
iris_set=iris[iris.species=='Iris-setosa']
seaborn.distplot(iris_vir['petal_length']).set(ylim=(0,5))
```

```python
g=seaborn.FacetGrid(iris,hue='species')   #gsb
g.map(seaborn.distplot,'petal_width')     #gsb
```

```python
seaborn.regplot(x='petal_width',y='petal_length',data=iris)
```

```python
g=seaborn.FacetGrid(iris,hue='species') 
g.set(xlim=(0,2.5))#gsb
g.map(seaborn.regplot,'petal_width','petal_length')
```
